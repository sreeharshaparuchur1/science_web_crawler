```python
import csv 
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
import urllib.request as url
import re
import requests
```

```python
Open_data = url.urlopen('https://bmccancer.biomedcentral.com/articles').read()
Soup_data = bs(Open_data,'lxml')
# Parsing through LXML
pg_range_text = (Soup_data.find_all('p',{'class':"u-text-sm u-reset-margin"}))[0].text
page_range = [int(s) for s in pg_range_text.split() if s.isdigit()]
link_pages = []
for pgnum in range(page_range[0],page_range[1] + 1):
    print(pgnum)
    link = 'https://bmccancer.biomedcentral.com/articles?searchType=journalSearch&sort=PubDate&page='+str(pgnum)
    Open = url.urlopen(link).read()
    Soup = bs(Open,'lxml')
    link_data = Soup.find_all('a',{'data-test':"title-link"})
    for j in link_data:
        page_link = 'https://bmccancer.biomedcentral.com' + j['href']
        link_pages.append(page_link)
        print(page_link)
   
```

```python
#open primary links and get names and links
```

```python
%%time
name_email ={} 
import time 
c=0
for link in link_pages:
    time.sleep(5)
    Open_data = url.urlopen(link).read() # Opening and parsing the link to get the author names.
    Soup_data = bs(Open_data,'lxml')
    Form_data = Soup_data.find_all('a', {'class':"c-article-button"})
    links = [] # empty list for every link. 
    for j in Form_data:
        links.append(j.attrs['href'])
    for k in links:
        name = (' '.join(k.split('3')[1].split('+'))).strip('"')
        try:
            Open = url.urlopen(k).read()
            Soup_data = bs(Open,'lxml') 
            Form_data_pdf = Soup_data.find_all('a', {'data-test':"pdf-link"})
            #Retrieve the email address of the author. 
            pdf_link = 'https:' + Form_data_pdf[0].attrs['href']
            print(pdf_link, name)
            text = requests.get(pdf_link)
            address = re.findall(r"[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.com", text.text, re.I) # gets the email address from the text. 
            if len(address) > 0:
                name_email[name]=address
        except:
            continue
    c+=1
 
```
